```python
import folium
import json

us_map = folium.Map([43, -100], zoom_start=4)
geo_json = json.load(open('./us-counties.json'))

def fillColor(feature):
    fips_code = feature['id']
    # can use FIPS County Code to map to a color
    return 'green' if 'e' in feature['properties']['name'].lower() else '#ffff00'

folium.GeoJson(geo_json,
               style_function=lambda feature: {
                    'fillColor': fillColor(feature),
                    'color': 'black',
                    'weight': 2,
                    'dashArray': '5, 5'
               }).add_to(us_map)
```
