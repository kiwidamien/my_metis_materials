```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

%matplotlib inline
def func(x):
    return np.exp(-x**2/2)


a, b = 2, 9  # integral limits

fig, ax = plt.subplots()

def plt_normal(mean = 0, linetype='r'):
    x = np.linspace(-3 + mean, 3 + mean)
    y = func(x - mean)

    plt.plot(x, y, linetype, linewidth=2)
    plt.ylim(ymin=0.0, ymax=1.0)

plt_normal(3)
plt_normal(6, 'b--')
def shade(mean, start, stop, facecolor='0.9'):
    # Make the shaded region
    ix = np.linspace(start, stop)
    iy = func(ix - mean)
    verts = [(start, 0)] + list(zip(ix, iy)) + [(stop, 0)]
    poly = Polygon(verts, color=facecolor, alpha=0.5)
    ax.add_patch(poly)

shade(3, 5, 6, 'r')
shade(6, 5, 10, 'b')

plt.figtext(0.9, 0.05, 'Difference in Proportions')
#plt.figtext(0.1, 0.9, '$y$')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')

ax.set_xticks((3,5))
ax.set_xticklabels((0,'crit',))
ax.set_yticks([])

plt.show()
```
