```python

def clean_text(text):
    
    text = text.encode('ascii', 'ignore').decode('utf-8')
    
    return [word for word in nltk.word_tokenize(text.lower()) if word[0] in string.letters and word not in stopwords]
```

```python

print("Original text:" + df.iloc[0,0])
print("Cleaned text:" + str(clean_text(df.iloc[0,0])))
```

```python
def process_text(all_gender_text):
    return nltk.FreqDist(all_gender_text)
```

```python
freq_dist = {}
for name, group in df.groupby('is_female'):
    author_text = group[0].str.cat(sep=' ')
    freq_dist[name] = process_text(clean_text(author_text))
```

```python
freq_dist.keys()
```

```python
import numpy as np

def bag_of_words(text, epsilon = 10**-9):
    cleaned = clean_text(text)  # get an array of words
    
    cumulative = {}
    for author in freq_dist:
        score = np.log([max(freq_dist[author].freq(word), epsilon) for word in cleaned]).sum()
        cumulative[author] = score
    return sorted([(value, author) for author, value in cumulative.items()], reverse=True)
```

```python
bag_of_words('Hello world')
```
